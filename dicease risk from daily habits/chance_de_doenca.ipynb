# Tratamento das colunas com dados ausentes

### Colunas do tipo categórico
* `alcohol_consumption`, `exercise_type`, `caffeine_intake`

```python
# Substituindo valores categóricos ausentes por 'unknown'
col_to_replace = ['alcohol_consumption', 'exercise_type', 'caffeine_intake']

for col in col_to_replace:
    df[col] = df[col].fillna('')
```

```python
for col in col_to_replace:
    df[col] = df[col].replace('', 'Unknown')
```

### Colunas do tipo quantitativo
* `blood_pressure`, `heart_rate`, `insulin`, `daily_steps`, `income`, `gene_marker_flag`

```python
quantitative_columns= ['blood_pressure', 'heart_rate', 'insulin', 'daily_steps', 'income', 'gene_marker_flag']

print(df[quantitative_columns].median())
```

```python
avg_col = []
for col in quantitative_columns:
    avg = df[col].mean().round(6)
    avg_col.append(avg)

print(avg_col)
```

A mediana é uma boa estimativa de valor esperado quando os dados são altamente dispersos. Isso ocorre porque ela não é tão distorcida por valores atípicos quanto a média.

```python
df['gene_marker_flag'] = df['gene_marker_flag'].fillna(avg_col[-1])
avg_col.pop()
quantitative_columns.pop()
print(df['gene_marker_flag'].unique())
```

```python
print(avg_col)
print(quantitative_columns)
```

```python
for i, col in enumerate(quantitative_columns):
    df[col] = df[col].fillna(avg_col[i])
```
